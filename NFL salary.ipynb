```python
import pandas as pd
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel("nfl_salary.xlsx")
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
# sum of missing values of each columns
df.isnull().sum()
```

```python
# replace it with mean
df.Quarterback.fillna((df.Quarterback.mean()), inplace = True)
df['Special Teamer'].fillna((df['Special Teamer'].mean()), inplace = True)
df.describe()
```

```python
year_2011 = df.loc[(df["year"] == 2011)]
year_2011.head()
```

```python
y_2011 = pd.pivot_table(df, columns='year', values= ['Cornerback','Defensive Lineman','Linebacker','Offensive Lineman','Quarterback','Running Back','Safety','Special Teamer','Tight End','Wide Receiver'], 
                        aggfunc= np.mean)
y_2011.reset_index(inplace=True)
y_2011.columns = ['Position', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']
y_2011
```

```python
red_patch = mpatches.Patch(color='red', label='2016')
blue_patch = mpatches.Patch(color='blue', label='2017')
green_patch = mpatches.Patch(color = 'green', label = '2018')


f, ax = plt.subplots(figsize=(10, 10))
p1 = sns.pointplot(x = 'Position', y = '2016',
                 color = 'red',data  = y_2011)
p2 = sns.pointplot(x = 'Position', y = '2017',
                 color = 'blue',data  = y_2011)
p3 = sns.pointplot(x = 'Position', y = '2018',
                 color = 'green',data  = y_2011)
plt.title('Salary Changes of Different Position From 2016 - 2018', size = 15)
plt.xlabel("Postion", size = 15)
plt.ylabel("Mean Salary", size = 15)
p3.set_xticklabels(p3.get_xticklabels(), rotation=90, size = 10)
plt.legend(handles=[red_patch, blue_patch, green_patch])
plt.show()
```
